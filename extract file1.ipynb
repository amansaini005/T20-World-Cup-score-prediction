```python
import numpy as np
import pandas as pd
import matplotlib as plt
from yaml import safe_load
import seaborn as sns
import os
from tqdm import tqdm
```

## Extract yaml file Path from os module

```python
filename =[]
for file in os.listdir('T20'):
    filename.append(os.path.join('T20', file))
```

```python
filename[:5]
```

## use safe _load to handle yaml file

```python
cric = pd.DataFrame()
counter = 1
for file in tqdm(filename):
    with open(file, 'r') as f:
        df = pd.json_normalize(safe_load(f))
        df['match_id'] = counter
        cric = cric.append(df)
        counter += 1
        
cric        
```

```python
count = 1
delivery = pd.DataFrame()
for index, row in matches.iterrows():
    if count in [45, 75, 100, 150, 175, 180, 260, 450, 650, 780, 1000]:
        count += 1
        continue
    count += 1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for k in ball.keys():
            match_id.append(count)
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(k)
            batsman.append(ball[k]['batsman'])
            bowler.append(ball[k]['bowler'])
            runs.append(ball[k]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[k]['wicket']['player_out'])
               
            except:
                player_of_dismissed.append('0')
             
    loop_df = pd.DataFrame({
        
            'match_id': match_id,
            'batting_team': batting_team,
            'teams': teams,
            'ball': ball_of_match,
            'batsman': batsman,
            'bowler': bowler,
            'runs': runs,
            'city': city,
            'venue': venue,
            'player_of_dismissed': player_of_dismissed
    })        
    
    delivery = delivery.append(loop_df)
                                    
```

```python
delivery
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery['bowling_team'] = delivery.apply(bowl, axis = 1)
```
